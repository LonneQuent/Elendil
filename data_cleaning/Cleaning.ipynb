```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_customers = pd.read_csv("olist_customers_dataset.csv")
df_sellers = pd.read_csv("olist_sellers_dataset.csv")
df_order_reviews = pd.read_csv("olist_order_reviews_dataset.csv")
df_order_items = pd.read_csv("olist_order_items_dataset.csv")
df_products = pd.read_csv("olist_products_dataset.csv")
df_category_name_translation = pd.read_csv("product_category_name_translation.csv")
df_orders = pd.read_csv("olist_orders_dataset.csv")
df_order_payments = pd.read_csv("olist_order_payments_dataset.csv")
```

```python
df_order_payments = df_order_payments.drop(['payment_sequential','payment_installments','payment_value'], axis=1)
```

```python
list_df = [df_customers, df_sellers, df_order_reviews, df_order_items, df_products, df_category_name_translation ,df_orders, df_order_payments]
```

```python

merged_df = pd.merge(df_order_items, df_orders, on='order_id', how='left')
print(merged_df.shape)

merged_df2 = pd.merge(merged_df, df_customers, on='customer_id', how='left')
print(merged_df2.shape)

merged_df3 = pd.merge(merged_df2, df_products, on='product_id', how='left')
print(merged_df3.shape)

merged_df5 = pd.merge(merged_df3, df_category_name_translation, on='product_category_name', how='left')
print(merged_df5.shape)

merged_df4 = pd.merge(merged_df5, df_order_payments, on='order_id', how='left')
print(merged_df4.shape)

df_final = pd.merge(merged_df4, df_order_reviews, on='order_id', how='left')
print(df_final.shape)
```

# Netoyage + analyse des donn√©es

```python
df_final = df_final.drop(['product_category_name', 'order_approved_at', 'seller_id','shipping_limit_date', 'customer_unique_id', 'customer_zip_code_prefix', 'product_name_lenght', 'product_description_lenght', 'product_photos_qty', 'product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm', 'review_id', 'review_comment_title', 'review_creation_date', 'review_answer_timestamp'], axis=1)
```

```python
df_final = df_final.rename(columns = {'product_category_name_english': 'product_category_name'})
```

```python
df_final =df_final.drop_duplicates()
```

```python
df_final.shape
```

```python
df_final['payment_type'] = df_final['payment_type'].replace({'boleto':'cash', 'voucher': "gift_card"})
```

```python
state_name = {
"AC" : "Acre",
"AL" : "Alagoas",
"AM" : "Amazonas",
"AP" : "Amapa",
"BA" : "Bahia",
"CE" : "Ceara",
"DF" : "Distrito Federal",
"ES" : "Espirito Santo",
"GO" : "Goias",
"MA" : "Maranhao",
"MG" : "Minas Gerais",
"MS" : "Mato Grosso do Sul",
"MT" : "Mato Grosso",
"PA" : "Para",
"PB" : "Paraiba",
"PE" : "Pernambuco",
"PI" : "Piaui",
"PR" : "Parana",
"RJ" : "Rio de Janeiro",
"RN" : "Rio Grande do Norte",
"RO" : "Rondonia",
"RR" : "Roraima",
"RS" : "Rio Grande do Sul",
"SC" : "Santa Catarina",
"SE" : "Sergipe",
"SP" : "Sao Paulo",
"TO" : "Tocantins"
}
```

```python
df_final['customer_state'] = df_final['customer_state'].map(state_name)
```

```python
df_final['date_only'] = pd.to_datetime(df_final['order_date']).dt.date
```

```python
df_final.head()
```

```python
df_final = df_final.rename(columns={"order_purchase_timestamp": "order_date",
                                        "order_delivered_carrier_date": "shipped_date",
                                        "order_delivered_customer_date": "delivery_date"})
```

```python
dt_columns = ['order_date', 'shipped_date', 'delivery_date']

for column in dt_columns:
  df_final[column] = pd.to_datetime(df_final[column])
```

```python
df_panier_group = df_final.groupby('order_id').agg({
    'order_item_id': 'max',  # Garde le order_item_id max
    'product_id': lambda x: list(x),  # Cr√©e une liste de tous les produits
    'price': 'sum',  # Somme de price
    'freight_value': 'first',  # Garde la premi√®re valeur de freight_value
    'product_category_name': lambda x: list(x)
}).reset_index()

df_panier_group['total_price'] = df_panier_group['price'] + df_panier_group['freight_value']
df_panier_group['date'] = df_final['date_only']
df_panier_group['customer_id'] = df_final['customer_id']
df_panier_group = df_panier_group.rename(columns={"order_item_id":"nb_article_panier"})
```

```python
nb_moyen_article_par_panier = df_panier_group['nb_article_panier'].sum()/len(df_panier_group['order_id'])
nb_moyen_article_par_panier
```
